## Way too long words
If the word length is greater than 4, then we use special rules to shorted the word length.
We take the first and last word and in between we use the length of words.

Ex: \
IP: internationalization\
OP: i18n\
Explaination: count of words between i and last letter n is 18


```python
string=input()
if len(string) > 4:
    res = ""
    res += string[0]
    count = 0

    for i in range(1, len(string)-1):
        count += 1

    res += str(count)
    res += string[-1]

    print(res)

else:
    print(string)
```
