```python
arr = [10, 8, 2, 99, 22]
def selectionSort(arr):
    for i in range(len(arr)-1):
        min_ind = i
        for j in range(i+1, len(arr)):
            print(arr)
            if arr[j] < arr[min_ind]:
                min_ind = j

        arr[i], arr[min_ind] = arr[min_ind], arr[i]

    return arr

selectionSort(arr)
```

```python
def lomutoPartition(arr, l, h):
    pivot = arr[h]
    i = l - 1

    for j in range(l, h):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[h] = arr[h], arr[i+1]

    return i+1

def qSort(arr, l, h):
    if l < h:
        p = lomutoPartition(arr, l, h)
        qSort(arr, l, p-1)
        qSort(arr, p+1, h)


arr = [10, 5, 2, 3, 19, 22]
qSort(arr, 0, len(arr)-1)
```

```python
def DFS(adj, s):
    visited = [False] * len(adj)
    DFS_trav(adj, s, visited)

def DFS_trav(adj, s, visited):
    visited[s] = True

    for i in adj[s]:
        if visited[i] == False:
            DFS_trav(adj, i, visited)
```
