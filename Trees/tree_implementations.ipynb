```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def inorder(root):
    if root != None:
        inorder(root.left)
        print(root.data)
        inorder(root.right)

def preorder(root):
    if root != None:
        print(root.data)
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root != None:
        postorder(root.left)
        postorder(root.right)
        print(root.data)
```

```python
def height(root):
    if root == None:
        return 0
    
    lr = height(root.left)+1
    rh = height(root.right)+1
    return max(lr, rh)+1
```

```python
def printKthNodes(root, k):
    if k == 0:
        return root.data
    
    if root == None:
        return None
    
    else:
        printKthNodes(root.left, k-1)
        printKthNodes(root.right, k-1)
```

```python
def levelOrderTraversal(root):
    if root == None:
        return
    q = []
    q.append(root)

    while len(q)>0:
        res = q.pop(0)
        print(res.data)

        if res.left != None:
            q.append(res.left)

        if res.right != None:
            q.append(res.right)
```

```python
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.left.right = Node(50)

levelOrderTraversal(root)
```

```python
def maximumElem(root):
    if root == None:
        return float('-inf')
    
    res = root.data

    lef = maximumElem(root.left)
    rig = maximumElem(root.right)

    if lef > res:
        res = lef

    if rig > res:
        res = lef

    return res

    
```
