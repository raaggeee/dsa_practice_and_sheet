```python
#Binary Tree
## Element on the left of the binary tree is smaller than the root
## Element on the right of the binary tree is greater than the root
```

```python
##Normal recusrive way
def searchBST(root, val):
    if root == None:
        return False
    
    if root.data == val:
        return True
    
    searchBST(root.left, val)
    searchBST(root.right, val)
```

```python
##Since binary tree has left element smaller than root and right elements greater than root. 
# We can comapre and recrusively go to the correct side rather than traversing the whole tree
def searchBST1(root, val):
    if root == None:
        return False
    
    if root.data == val:
        return True
    
    if root.data > val:
        return searchBST(root.left, val)

    if root.data < val:
        return searchBST(root.right, val)
```

```python
## Iterative way to traverse
def searchBST(root, val):
    while root != None:
        if root.data == val:
            return True
        
        elif root.data > val:
            root = root.left

        else:
            root = root.right

    return False
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = Node(40)
root.left = Node(20)
root.right = Node(80)
root.left.left = Node(8)
root.left.right = Node(36)
root.right.left = Node(60)
root.right.right = Node(100)

searchBST1(root, 4)
```

```python
## Iterative way to insert
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insertBST2(root, val):
    if root == None:
        return Node(val)
    
    curr = root

    while curr != None:
        prevCurr = curr

        if curr.data == val:
            return root 
        
        elif curr.data > val:
            curr = curr.left

        elif curr.data < val:
            curr = curr.right

    if val < prevCurr.data:
        prevCurr.left = Node(val)

    elif val > prevCurr.data:
        prevCurr.right = Node(val)

    return root
```

```python
def getSucc(root):
    while root.left != None:
        root = root.left

    return root.data

def delNode(root, val):
    if root == None:
        return None
    
    if root.data > val:
        root.left = delNode(root.left, val)

    elif root.data < val:
        root.right = delNode(root.right, val)

    else:
        if root.left == None:
            return root.right
        
        elif root.right == None:
            return root.left
        
        else:
            succ = getSucc(root)
            root.key = succ
            root.right = delNode(root.right, succ)

    return root
```
